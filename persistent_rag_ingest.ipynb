```python
from ingest import load_faq_data
documents = load_faq_data()
```

```python
docs_llm = [doc for doc in documents if doc["course"] == "llm-zoomcamp"]
print(docs_llm)
#print(f"LLM Zoomcamp: {len(docs_llm)} documents")
```

```python
import time
from sqlitesearch import TextSearchIndex

index = TextSearchIndex(
    text_fields=["question", "section", "answer"],
    keyword_fields=["course"],
    id_field="doc_id",      # <-- use the existing id so if it's re-run, it will update existing records instead of creating duplicates
    db_path="faq.db"
)

print("Beginning files number: " + str(index.count()))

for doc in docs_llm:
    index.add(doc)
    print(f"""Added: {doc["question"][:60]}...""")
    time.sleep(0.5)

print("Ending files number: " + str(index.count()))

index.close()
print("Done. Index saved to faq.db")
```
